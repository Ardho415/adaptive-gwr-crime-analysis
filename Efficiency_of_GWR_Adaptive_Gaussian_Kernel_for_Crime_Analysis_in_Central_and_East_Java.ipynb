```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import libpysal as ps
from esda.moran import Moran_Local
from mgwr.gwr import GWR
from mgwr.sel_bw import Sel_BW

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.float_format', '{:.6f}'.format)

# Load Data
try:
    data = pd.read_csv('data_jateng.csv')
    coords = list(zip(data['latitude'], data['longitude']))
    y = data['y'].values.reshape(-1, 1)
    X_vars = data[['x1', 'x2', 'x3', 'x4']].values
    X = np.hstack([np.ones((len(y), 1)), X_vars])
    kota_kab = data['kota_kabupaten'].values
    print("Data berhasil dimuat.")
except FileNotFoundError:
    print("File 'data_jateng.csv' tidak ditemukan.")
```

 **Analisis LISA**

```python
print("TABEL 1: HASIL UJI LOCAL MORAN'S I (LISA) PADA RESIDUAL OLS")

# 1. OLS Model
model_ols = sm.OLS(y, X).fit()
residuals = model_ols.resid

# 2. LISA Test
w_knn = ps.weights.KNN.from_array(coords, k=5)
w_knn.transform = 'R'
lisa = Moran_Local(residuals, w_knn, permutations=999)

# Membuat DataFrame Hasil LISA
lisa_results = pd.DataFrame({
    'Kota/Kabupaten': kota_kab,
    'Residual OLS': residuals,
    'LISA Index (Is)': lisa.Is,
    'p-value': lisa.p_sim,
    'Cluster Type': lisa.q # 1:HH, 2:LH, 3:LL, 4:HL
})

# Filter hanya yang signifikan (p-value < 0.05)
lisa_signifikan = lisa_results[lisa_results['p-value'] < 0.05].copy()

# Mapping kuadran
quadrant_map = {1: 'High-High', 2: 'Low-High', 3: 'Low-Low', 4: 'High-Low'}
lisa_signifikan['Label Cluster'] = lisa_signifikan['Cluster Type'].map(quadrant_map)

from IPython.display import display
print(f"Ditemukan {len(lisa_signifikan)} lokasi dengan autokorelasi spasial signifikan:")
display(lisa_signifikan[['Kota/Kabupaten', 'Residual OLS', 'LISA Index (Is)', 'p-value', 'Label Cluster']])
```

**Pemilihan Kernel**

```python
print("TABEL 2: PERBANDINGAN PERFORMA MODEL GWR BERDASARKAN KERNEL")

kernel_list = ['bisquare', 'gaussian', 'exponential']
comparison_data = []

best_aicc = float('inf')
best_kernel = ''
best_bw = 0

for k_name in kernel_list:
    # Cari Bandwidth
    selector = Sel_BW(coords, y, X, kernel=k_name, fixed=False, spherical=True)
    bw = selector.search(bw_min=2)

    # Fit Model
    model = GWR(coords, y, X, bw=bw, kernel=k_name, fixed=False, spherical=True)
    results = model.fit()

    # Simpan Data
    comparison_data.append({
        'Jenis Kernel': f"Adaptive {k_name.capitalize()}",
        'Bandwidth (Tetangga)': int(bw),
        'AICc': results.aicc,
    })

    # Cek AICc Terbaik
    if results.aicc < best_aicc:
        best_aicc = results.aicc
        best_kernel = k_name
        best_bw = bw

df_comparison = pd.DataFrame(comparison_data).sort_values(by='AICc')
display(df_comparison)

print(f"\nModel terbaik adalah kernel '{best_kernel}' dengan AICc terendah.")
```

**Estimasi Parameter Lokal**

```python
print(f"TABEL 3: ESTIMASI PARAMETER GWR (KERNEL: {best_kernel.upper()}) SETIAP LOKASI")

# Menggunakan Kernel Gaussian
model_final = GWR(coords, y, X, bw=best_bw, kernel=best_kernel, fixed=False, spherical=True)
results_final = model_final.fit()

params = results_final.params

df_params = pd.DataFrame(params, columns=['Intercept', 'b1 (x1)', 'b2 (x2)', 'b3 (x3)', 'b4 (x4)'])
df_params.insert(0, 'Kota/Kabupaten', kota_kab)

df_params['Local R2'] = results_final.localR2

display(df_params)

# Simpan ke csv
df_params.to_csv('Hasil_GWR.csv', index=False)
print("Tabel telah disimpan sebagai 'Hasil_GWR.csv'")
```

# **Hasil dan Pembahasan**

**OLS VS GWR**

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from IPython.display import display, HTML

print("--- ANALISIS KOMPARASI: GLOBAL (OLS) vs LOKAL (GWR) ---")

# 1. OLS
model_ols = sm.OLS(y, X).fit()
y_pred_ols = model_ols.fittedvalues.reshape(-1, 1)

# 2. Ambil Hasil GWR
y_pred_gwr = results_final.predy.reshape(-1, 1)
y_asli = y.reshape(-1, 1)

# 3. Fungsi untuk menghitung AICc
# Rumus AICc = -2*LL + 2k + (2k(k+1))/(n-k-1)
def calc_ols_aicc(model_ols_res):
    ll = model_ols_res.llf
    n = model_ols_res.nobs
    k = model_ols_res.df_model + 1 # +1 untuk intercept
    aicc = -2 * ll + 2 * k + (2 * k * (k + 1)) / (n - k - 1)
    return aicc

# 4. Hitung Metrik Perbandingan
metrics = {
    'Indikator': ['AICc (Efisiensi)', 'Adjusted R-squared', 'RMSE (Error)', 'MAPE (Akurasi)'],

    # Metrik OLS
    'Model Global (OLS)': [
        calc_ols_aicc(model_ols),
        model_ols.rsquared_adj,
        np.sqrt(mean_squared_error(y_asli, y_pred_ols)),
        mean_absolute_percentage_error(y_asli, y_pred_ols) * 100
    ],

    # Metrik GWR
    'Model Lokal (GWR)': [
        results_final.aicc,
        results_final.adj_R2,
        np.sqrt(mean_squared_error(y_asli, y_pred_gwr)),
        mean_absolute_percentage_error(y_asli, y_pred_gwr) * 100
    ]
}

df_compare = pd.DataFrame(metrics)

# Kolom Keterangan
def cek_improvement(row):
    ols = row['Model Global (OLS)']
    gwr = row['Model Lokal (GWR)']

    # Untuk R2, makin besar makin baik
    if 'R-squared' in row['Indikator']:
        diff = gwr - ols
        return f"Naik +{diff:.3f}" if diff > 0 else "Turun"
    # Untuk AICc, RMSE, MAPE, makin kecil makin baik
    else:
        diff = ols - gwr
        return f"Turun -{diff:.3f}" if diff > 0 else "Naik (Memburuk)"

df_compare['Perbaikan (GWR vs OLS)'] = df_compare.apply(cek_improvement, axis=1)

pd.set_option('display.float_format', '{:.4f}'.format)
display(HTML(df_compare.to_html(index=False)))

print("\nKESIMPULAN:")
if df_compare.iloc[0, 2] < df_compare.iloc[0, 1]:
    print("Model GWR lebih baik daripada OLS di semua indikator.")
    print("   (AICc lebih rendah, R-squared lebih tinggi, dan Error lebih kecil).")
else:
    print("Perhatikan hasil: Apakah GWR benar-benar menurunkan AICc secara signifikan?")
```

```python
import pandas as pd
from IPython.display import display, HTML

print("\nTABEL 4: RINGKASAN STATISTIK ESTIMASI PARAMETER LOKAL")

# Membuat Tabel Statistik
cols_to_analyze = ['Intercept', 'b1 (x1)', 'b2 (x2)', 'b3 (x3)', 'b4 (x4)', 'Local R2']
summary_stats = df_params[cols_to_analyze].describe().loc[['min', 'max', 'mean', 'std']]

final_summary = summary_stats.transpose()
final_summary.columns = ['Minimum', 'Maksimum', 'Rata-rata', 'Standar Deviasi']
display(final_summary)

print("\nTABEL 5: RINGKASAN VARIASI SPASIAL PARAMETER MODEL GWR")

# 4. Variabel
vars_to_analyze = {
    'Local R2': 'Local R^2',
    'b1 (x1)': 'Koefisien x1',
    'b2 (x2)': 'Koefisien x2',
    'b3 (x3)': 'Koefisien x3',
    'b4 (x4)': 'Koefisien x4'
}

summary_data = []

for col_name, label_paper in vars_to_analyze.items():
    # Cari Nilai Max
    idx_max = df_params[col_name].idxmax()
    kota_max = df_params.loc[idx_max, 'Kota/Kabupaten']
    val_max = df_params.loc[idx_max, col_name]

    # Cari Nilai Min
    idx_min = df_params[col_name].idxmin()
    kota_min = df_params.loc[idx_min, 'Kota/Kabupaten']
    val_min = df_params.loc[idx_min, col_name]

    # Sifat Pengaruh
    if 'R^2' in label_paper:
        sifat = "Fit Model"
    else:
        if val_min > 0 and val_max > 0:
            sifat = "Positif Konsisten"
        elif val_min < 0 and val_max < 0:
            sifat = "Negatif Konsisten"
        else:
            sifat = "Campuran (+ dan -)"

    summary_data.append({
        'Parameter': label_paper,
        'Lokasi Nilai Maksimum': kota_max,
        'Nilai Max': f"{val_max:.4f}",
        'Lokasi Nilai Minimum': kota_min,
        'Nilai Min': f"{val_min:.4f}",
        'Sifat Pengaruh': sifat
    })

df_summary_final = pd.DataFrame(summary_data)
display(HTML(df_summary_final.to_html(index=False)))
```
